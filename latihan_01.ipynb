```python
mahasiswa = ["ryan", "maha", "mahesa", "reno", "zizu"]
print("list mahasiswa: ")
print(mahasiswa)

```

Tambahan 1 nama baru dengan append

```python
mahasiswa.append("nyoman")
print("\nSetelah di tambahkan nama baru:")
print(mahasiswa)
```

hapus 1 elemen menggunakan remove

```python
mahasiswa.remove:("ryan")
print("\nSetelah menghapus nama ryan:")
print(mahasiswa)
```
